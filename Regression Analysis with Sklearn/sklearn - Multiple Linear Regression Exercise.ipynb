## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

## Load the data

```python
est_data = pd.read_csv('real_estate_price_size_year.csv')
est_data.head()
```

```python
est_data
```

## Create the regression

### Declare the dependent and the independent variables

```python
y = est_data['price']
x = est_data[['size','year']]
```

### Regression

```python
reg = LinearRegression()
reg.fit(x,y)
```

### Find the intercept

```python
reg.intercept_
```

### Find the coefficients

```python
reg.coef_
```

### Calculate the R-squared

```python
reg.score(x,y)
```

### Calculate the Adjusted R-squared

```python
def adjusted_r_sq(x, y):
    r2 = reg.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r_sq = 1-(1-r2)*(n-1)/(n-p-1)
    
    return adjusted_r_sq
```

```python
adjusted_r_sq(x, y)
```

### Making predictions

Find the predicted price of an apartment that has a size of 750 sq.ft. from 2009.

```python
reg.predict([[750,2009]])
```

### Calculate the univariate p-values of the variables

```python
from sklearn.feature_selection import f_regression
f_regression(x,y)[1].round(2)
```

### Create a summary table with your findings

```python
summary = pd.DataFrame(data = x.columns.values, columns=['Features'])
summary ['Coefficients'] = reg.coef_
p = f_regression(x,y)[1].round(2)
summary ['p-values'] = p
summary
```
